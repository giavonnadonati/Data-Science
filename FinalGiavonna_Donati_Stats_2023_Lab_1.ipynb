```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**8, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()
# first plot linear, 2nd quadratic, third cubic because of powers given to x
```

```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^');
#letters in commas defining color, red is linear, blue in quadratic, green is cubic
```

```python
names = ['cat', 'dog', 'bird']
values = [15, 18, 6]

plt.figure(1, figsize=(9, 3))


plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names, values)


plt.suptitle('Categorical Plotting')
plt.show()
# names= gives name to the categories, values = gives values to respective names
#plt.bar gives bar graph, plt.scatter gives scatter, plt.plot gives line
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)

plt.plot(t, s)
plt.title(r'$\alpha_i > \beta_i$', fontsize=20)
plt.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=20)
plt.text(0.5, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',fontsize=15)
plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-10, 10, 1000)
y = 10*x + 7

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(xlabel='x', ylabel='y',
       title='Graph of y = 10x+7')
ax.grid()

fig.savefig("test.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-10, 10, 1000)
y = -3*x - 12

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(xlabel='x', ylabel='y',
       title='Graph of y = -3x - 12')
ax.grid()

fig.savefig("test.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()
ax = plt.axes()

x = np.linspace(-10, 10, 1000)
ax.plot(x, np.sin(x));


ax.set(xlabel='x', ylabel='y',
       title='Graph of y = sin(x)')
ax.grid()

fig.savefig("test.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np



fig = plt.figure()
ax = plt.axes()

x = np.linspace(-10, 10, 1000)
ax.plot(x, np.exp(x));


ax.set(xlabel='x', ylabel='y',
       title='Graph of y = e^x')
ax.grid()

fig.savefig("test.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np



fig = plt.figure()
ax = plt.axes()

x = np.linspace(-10, 10, 1000)
ax.plot(x, np.exp(-x**2));


ax.set(xlabel='x', ylabel='y',
       title='Graph of y = e^-x^2')
ax.grid()

fig.savefig("test.png")
```
