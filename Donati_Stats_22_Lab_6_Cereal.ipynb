```python
x = np.linspace(0,10,50)
#defining x as 50 guesses from 0 to 10
```

```python
noise = 10 * np.random.randn(50,)
#defining noise
```

```python
y = 3 * x + 7 + noise
```

```python
plt.plot(x,y,'.');
#plotting x and y
```

```python
m_guess = 1
b_guess = 1

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
#defining baseline guesses for m b and y
```

```python
loss = np.sum((y_guess - y)**2)
#defining loss
```
